```python
import numpy as np
import csv
import os
```

```python
def glue_data(file_name, idx, total, output):
    gluedata = 0
    for i in range(total):
        file_name1 = file_name + str(idx) + '_' + str(i) + '.csv'
        try:
            with open(file_name1, 'r', newline='') as f:
                lines = csv.reader(f)
                raw_data = []
                for line in lines:
                    raw_data.append(line)
            raw_data = np.array(raw_data).astype(np.float)
            #print(i)
            if i == 0:
                gluedata = raw_data
            else:
                smooth = (gluedata[:, -1] + raw_data[:, 1]) / 2
                gluedata[:, -1] = smooth
                raw_data[:, 1] = smooth
                gluedata = np.append(gluedata, raw_data, axis=1)
        except:
            continue
    #print(gluedata.shape)
    filename2 = output
    with open(filename2, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerows(gluedata)
        print("GLUE DONE!" + filename2)
```

```python
total = [60, 74, 75, 76, 77, 75, 77, 59, 59, 78, 80, 77, 81, 58, 77, 77, 75, 78, 78, 81, 78, 78, 76, 86, 76, 77, 78, 78, 75, 59, 75, 59, 76, 75, 75, 60, 75, 78, 65, 61, 78, 75, 81, 85, 75, 78, 57, 78, 76]
for i in range(49):
    idx = i+446
    try:
        outputname = "./Real_EEG/testing/Other/" + str(idx) + ".csv"
        glue_data("./Real_EEG/test/Other/", idx, total[i], outputname)
    except:
        continue
            
```
